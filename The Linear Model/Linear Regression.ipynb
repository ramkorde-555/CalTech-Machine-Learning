```python
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
X, y = datasets.make_regression(n_samples=200,#number of samples
                                      n_features=1,#number of features, 1 for visualization
                                      n_informative=1,#number of useful feature
                                      bias = 5, #bias in underlying model
                                      noise=15,#standard deviation of the guassian noise
                                      coef = False,
                                      random_state=2)
```

```python
X = np.insert(X, 0, 1.0, axis = 1)
```

```python
plt.ion() #interactive plot on
plt.plot(X[:, 1],y,'.',label='training data')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Plot on 2D plane')
```

```python
# The Linear Regression model

# computinng X_dagger, the psedo-inverse of X : X_d

X_pinv = np.linalg.inv(np.dot(X.T, X)) # inverse of (X transpose dot X)
X_d = np.dot(X_pinv, X.T)

# calculate weights using X_d : W = X_d dot y

W = np.dot(X_d, y)
W = W.reshape(-1,1)

# The final hypothesis W is given as the function
def h(x, W):
    x = np.insert(x, 0, 1.0, axis = 1)
    return np.dot(x, W)
```

```python
X_in1 = np.array([-3.69]).reshape(-1,1)
X_in2 = np.array([4.20]).reshape(-1,1)
```

```python
y_hat1 = h(X_in1, W)
y_hat2 = h(X_in2, W)
```

```python
plt.plot(X[:, 1],y,'.',label='training data')
plt.xlabel('X')
plt.ylabel('y')
plt.plot([np.squeeze(X_in1), np.squeeze(X_in2)], [np.squeeze(y_hat1), np.squeeze(y_hat2)])
plt.show()
```
